```python
graph = {
"A": ["B", "C"],
"B": ["D", "E"],
"C": ["F"],
"D": [],
"E": ["F"],
"F": []
}

def dfs_path(graph, start, goal, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []
    visited.add(start)
    path.append(start)
    if start == goal:
        return path
    for neighbor in graph.get(start, []):
        if neighbor not in visited:
            result = dfs_path(graph, neighbor, goal, visited, path)
            if result is not None:
                return result
    path.pop()
    return None

print(dfs_path(graph, "A", "F"))
```

```python
from collections import deque
import time

def bfs_shortest_path(graph, start, goal):
    start_time = time.time()

    queue = deque([[start]])
    visited = set()
    visited_count = 0

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node in visited:
            continue
        visited.add(node)
        visited_count += 1

        if node == goal:
            end_time = time.time()
            return path, visited_count, end_time - start_time

        for neighbor in graph.get(node, []):
            new_path = list(path)
            new_path.append(neighbor)
            queue.append(new_path)

    return None, visited_count, time.time() - start_time


import heapq
import time

def astar(graph, heuristic, start, goal):
    start_time = time.time()

    open_set = []
    heapq.heappush(open_set, (heuristic[start], 0, start, [start]))
    visited = set()
    visited_count = 0

    while open_set:
        f, g, current, path = heapq.heappop(open_set)

        if current == goal:
            end_time = time.time()
            return path, visited_count, end_time - start_time

        if current in visited:
            continue
        visited.add(current)
        visited_count += 1

        for neighbor, cost in graph[current].items():
            if neighbor not in visited:
                new_g = g + cost
                new_f = new_g + heuristic[neighbor]
                heapq.heappush(open_set, (new_f, new_g, neighbor, path + [neighbor]))

    return None, visited_count, time.time() - start_time

graph_bfs = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D', 'E'],
    'D': [],
    'E': ['F', 'G'],
    'F': [],
    'G': []
}

graph_astar = {
    'A': {'B': 1, 'C': 1},
    'B': {'D': 1},
    'C': {'D': 1, 'E': 1},
    'D': {},
    'E': {'F': 1, 'G': 1},
    'F': {},
    'G': {}
}
heuristic = {
  'A': 4,
  'B': 3,
  'C': 3,
  'D': 2,
  'E': 2,
  'F': 0,
  'G': 1
}

print("BFS: ")
print(bfs_shortest_path(graph_bfs, 'A', 'F'))
print("A*: ")
print(astar(graph_astar, heuristic, 'A', 'F'))
```
