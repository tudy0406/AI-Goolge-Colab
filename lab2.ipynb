# Seaborn Titanic Dataset – Feature Description

Each row represents a **passenger**, and each column represents a **feature describing that passenger**.

---

# Variables

## survived
- **Type:** Binary (0 or 1)
- **Description:** Indicates whether the passenger survived the Titanic disaster.

| Value | Meaning |
|------|--------|
| 0 | Did not survive |
| 1 | Survived |

This is typically used as the **target variable** in machine learning models.

---

## pclass
- **Type:** Ordinal categorical (1, 2, 3)
- **Description:** Passenger ticket class.

| Value | Meaning |
|------|--------|
| 1 | First class  |
| 2 | Second class |
| 3 | Third class

---

## sex
- **Type:** Categorical
- **Description:** Passenger gender.

Possible values:
- `male`
- `female`

---

## age
- **Type:** Numeric (float)
- **Description:** Passenger age in years.

Some values may be **missing**.

---

## sibsp
- **Type:** Integer
- **Description:** Number of **siblings or spouses** the passenger had aboard the Titanic.

---

## parch
- **Type:** Integer
- **Description:** Number of **parents or children** the passenger had aboard the Titanic.

---

## fare
- **Type:** Numeric (float)
- **Description:** Price paid for the ticket.

Higher fares are typically associated with **higher passenger classes**.

---

## embarked
- **Type:** Categorical code
- **Description:** Port where the passenger boarded the Titanic.

| Code | Port |
|-----|------|
| C | Cherbourg |
| Q | Queenstown |
| S | Southampton |

---

## class
- **Type:** Categorical
- **Description:** Text version of `pclass`.

Possible values:
- `First`
- `Second`
- `Third`

---

## who
- **Type:** Categorical
- **Description:** Passenger group classification used for analysis.

Possible values:
- `man`
- `woman`
- `child`

---

## adult_male
- **Type:** Boolean
- **Description:** Indicates whether the passenger is an **adult male**.

| Value | Meaning |
|------|--------|
| True | Adult male |
| False | Otherwise |

---

## deck
- **Type:** Categorical
- **Description:** Deck level where the passenger's cabin was located.

Possible values:
- `A`
- `B`
- `C`
- `D`
- `E`
- `F`
- `G`

Many entries are **missing**, because cabin numbers were not recorded for many passengers.

---

## embark_town
- **Type:** Categorical
- **Description:** Full name of the embarkation port.

Possible values:
- `Cherbourg`
- `Queenstown`
- `Southampton`

---

## alone
- **Type:** Boolean
- **Description:** Indicates whether the passenger travelled **alone**.

| Value | Meaning |
|------|--------|
| True | Passenger travelled alone |
| False | Passenger had family aboard |

---
